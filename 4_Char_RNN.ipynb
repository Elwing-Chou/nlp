```
with open("倚天屠龍記.txt", "r", encoding="utf-8") as f:
    text = f.read()[:100000]
```

```
n = len(text)
print("倚天屠龍記共有 {} 中文字".format(n))
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
import numpy as np
tok = Tokenizer(filters=None, char_level=True)
tok.fit_on_texts(text)
word_to_index = tok.word_index
word_to_index
```

```
index_to_word = {v:k for k, v in word_to_index.items()}
index_to_word
```

```
chars = [c for c in text if c in word_to_index]
charints = [word_to_index[c] for c in chars]
V = len(word_to_index)
print("包含了 {} 個獨一無二的字".format(V))
print("原文(10):", text[:10])
print("原文(10):", chars[:10])
print("轉換(10):", charints[:10])
```

```
def generate_data(corpus, window=10):
    times = len(corpus) // window
    result = []
    for i in range(times):
        start = i * window
        end = i * window + (window + 1)
        words = corpus[start:end]
        if len(words) == window + 1:
            x = words[:-1]
            y = words[1:]
            result.append((x, y))
    return result
g = generate_data(charints)
g
```

```
import random
def batch_data(corpus, batch=64, window=10):
    g = generate_data(corpus, window)
    random.shuffle(g)
    batchx = []
    batchy = []
    for x, y in g:
        batchx.append(np.array(x))
        batchy.append(np.array(y))
        if len(batchx) == batch:
            yield (np.array(batchx), np.array(batchy))
            batchx = []
            batchy = []
g = batch_data(charints)
for _ in range(2):
    print(g.__next__())
```

```
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
EMBEDDING_DIM = 128
RNN_UNITS = 64
BATCH_SIZE = 64
SEQ_LENGTH = 256
# 使用 keras 建立一個非常簡單的 LSTM 模型
model = Sequential()
# 詞嵌入層
# 將每個索引數字對應到一個高維空間的向量
model.add(Embedding(input_dim=V+1,
                    mask_zero=True,
                    output_dim=EMBEDDING_DIM,
                    batch_input_shape=(BATCH_SIZE, SEQ_LENGTH)))

# LSTM 層
# 負責將序列數據依序讀入並做處理
model.add(LSTM(units=RNN_UNITS,
               return_sequences=True))

# 全連接層
# 負責 model 每個中文字出現的可能性
model.add(Dense(V+1, activation="softmax"))
model.summary()
```

```
model.compile(loss="categorical_crossentropy", optimizer="adam")
```

```
from tensorflow.keras.utils import to_categorical
for epoch in range(1000):
    loss = 0.
    for x, y in batch_data(charints, BATCH_SIZE, SEQ_LENGTH):
        y_cat = to_categorical(y, V+1)
        l = model.train_on_batch(x, y_cat)
        loss = loss + l
    if (epoch + 1) % 50 == 0:
        print("Epoch:", epoch)
        print("Loss:", loss)
```

```
PRE_BATCH_SIZE = 1
# 使用 keras 建立一個非常簡單的 LSTM 模型
infer = Sequential()
# 詞嵌入層
# 將每個索引數字對應到一個高維空間的向量
infer.add(Embedding(input_dim=V+1,
                    mask_zero=True,
                    output_dim=EMBEDDING_DIM,
                    batch_input_shape=(PRE_BATCH_SIZE, None)))

# LSTM 層
# 負責將序列數據依序讀入並做處理
infer.add(LSTM(units=RNN_UNITS,
               return_sequences=True,
               stateful=True))

# 全連接層
# 負責 model 每個中文字出現的可能性
infer.add(Dense(V+1, activation="softmax"))
infer.summary()
infer.set_weights(model.get_weights())
```

```
import textwrap

sampled = [word_to_index["張"]]
for _ in range(512):
    if len(sampled) == 0:
        batch = np.zeros((1, 1))
        c = np.random.randint(1, V+1)
        batch[0, 0] = c
        print("起始字:", index_to_word[c])
        sampled.append(c)
    else:
        batch = np.zeros((1, 1))
        batch[0, 0] = sampled[-1]
    result = infer.predict_on_batch(batch).ravel()
    sample = np.random.choice(list(range(V+1)), p=result)
    sampled.append(sample)

article = "".join([index_to_word[c] for c in sampled])
print("\n".join(textwrap.wrap(article, 30)))
```
